```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import scipy.ndimage as ndi
from skimage import io, feature, color, filters, img_as_float
from matplotlib import pyplot as plt                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                          
import cv2
```

```python
# defining global variable path
image_path = "./data/train_data"
output_path = "./processed_data/train_data"

# image_path = "./data/test_data"
# output_path = "./processed_data/test_data"


def loadImages(path):
    # Put files into lists and return them as one list of size 4
    file_path = os.path.join(image_path, path)
    image_files = sorted([os.path.join(file_path, file)
         for file in os.listdir(file_path) if      file.endswith('.png')])
 
    return image_files
```

```python

# Display one image
def display_one(a, title1 = "Original"):
    plt.imshow(a, cmap=plt.get_cmap('gray')), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.show()
# Display two images
def display(a, b, title1 = "Original", title2 = "Edited"):
    plt.subplot(121), plt.imshow(a, cmap=plt.get_cmap('gray')), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(b, cmap=plt.get_cmap('gray')), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
```

```python
def crop_center(img,cropx,cropy):
    y,x, dim = img.shape
    startx = x//2-(cropx//2)
    starty = y//2-(cropy//2)    
    return img[starty:starty+cropy,startx:startx+cropx]
```

```python
def find_pix_dim(ct_img):
    """
    Get the pixdim of the CT image.
    A general solution that gets the pixdim indicated from the image dimensions. From the last 2 image dimensions, we get their pixel dimension.
    Args:
        ct_img: nib image

    Returns: List of the 2 pixel dimensions
    """
    pix_dim = ct_img.header["pixdim"] # example [1,2,1.5,1,1]
    dim = ct_img.header["dim"] # example [1,512,512,1,1]
    max_indx = np.argmax(dim)
    pixdimX = pix_dim[max_indx]
    dim = np.delete(dim, max_indx)
    pix_dim = np.delete(pix_dim, max_indx)
    max_indy = np.argmax(dim)
    pixdimY = pix_dim[max_indy]
    return [pixdimX, pixdimY] # example [2, 1.5]
```

```python
from skimage import measure
# , cv2.IMREAD_GRAYSCALE
raw_img = cv2.imread('./data/train_data/goat/talus_goat_CA_AC_01_MNHN_20210304-170247.png')


original_image = img_as_float(raw_img)
gray = color.rgb2gray(original_image)
plt.hist(original_image.ravel(), bins=20, range=(0.0, 1.0), ec='k')

contours = measure.find_contours(gray, 0.7)
k = 2
sigma = 2

fig, ax = plt.subplots()
ax.imshow(original_image, cmap=plt.cm.gray)

for contour in contours:
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
plt.show()

s1 = filters.gaussian(original_image, k*sigma)
s2 = filters.gaussian(original_image, sigma)
dog = s1 - s2
plt.figure(figsize=(5,5))
plt.imshow(dog)

sobel_ax0 = ndi.sobel(dog, axis=0)
sobel_ax1 = ndi.sobel(dog, axis=1)

# Calculate edge magnitude 
edges = np.sqrt(np.square(sobel_ax0) + np.square(sobel_ax1))
plt.figure(figsize=(5,5))
plt.imshow((edges * 255).astype(np.uint8), cmap='gray', vmax=75)
```

```python
img = cv2.imread('./data/train_data/goat/talus_goat_CA_AC_01_MNHN_20210304-170247.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gaussian = cv2.GaussianBlur(gray,(3,3),0)

#canny
img_canny = cv2.Canny(img,100,200)

#sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely


#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)

plt.figure(figsize=(5,5))
plt.imshow(img)
plt.title('Original Image')

plt.figure(figsize=(5,5))
plt.imshow(gray + img_canny)
plt.title("Canny")
plt.figure(figsize=(5,5))
plt.imshow(img_sobelx)
plt.title("Sobel X")
plt.figure(figsize=(5,5))
plt.imshow(img_sobely)
plt.title("Sobel Y")
plt.figure(figsize=(5,5))
plt.imshow(img_sobel)
plt.title("Sobel")
plt.figure(figsize=(5,5))
plt.imshow(img_prewittx)
plt.title("Prewitt X")
plt.figure(figsize=(5,5))
plt.imshow(img_prewitty)
plt.title("Prewitt Y")
plt.figure(figsize=(5,5))
plt.imshow(img_prewittx + img_prewitty)
plt.title("Prewitt")
```

### Create function to convert data

```python
import scipy.ndimage as ndi
import numpy as np
from skimage import io, feature, color, filters, img_as_float
from matplotlib import pyplot as plt
    
# Preprocessing
def processing(path):
    k = 2
    sigma = 2
    # loading image
    data = loadImages(path)
    # Getting 3 images to work with 
    img =  [cv2.imread(i, cv2.IMREAD_GRAYSCALE) for i in data]
    for i in range(len(img)):
        # basic way
        image = img[i]
        
        blur = cv2.GaussianBlur(image, (5, 5), 0)

#         display(image, gray, 'Original', 'GRAY')
        ret, thresh = cv2.threshold(blur, 244, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)



        # Further noise removal
        kernel = np.ones((3, 3), np.uint8)
        opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=5)

        # sure background area
        sure_bg = cv2.dilate(opening, kernel, iterations=3)

        # Finding sure foreground area
        dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 3)
        ret, sure_fg = cv2.threshold(dist_transform, 0.5 * dist_transform.max(), 255, 0)

        # Finding unknown region
        sure_fg = np.uint8(sure_fg)
        unknown = cv2.subtract(sure_bg, sure_fg)


        file_name = os.path.join(output_path, path , "image_" + str(i) +'.png')
        cv2.imwrite(file_name,sure_bg)
#         cv2.imshow('BGR Image',  np.array(markers, dtype = np.uint8 ))
    
```

### Apply on dataset

```python
processing("goat")
processing("mouflon")
processing("rupicapra")
processing("sheep")
```

### Rotation Top view and pseudo_RGB data gen

```python
#generate top images for bones
from PIL import Image
from PIL import ImageOps

from os import listdir
from os.path import isfile, join

mypath = "C:/Users/frafa/UCA/Case study/Read_3D_im/test_data/TOP_corrected/"
# mypath = "C:/Users/frafa/UCA/Case study/Read_3D_im/test_data/TOP/"
_2Dfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

count = 0
for _2Dpic in _2Dfiles:
    count += 1

    name = _2Dpic.split('.')[0]
    input_file = mypath + _2Dpic

    im = Image.open(input_file)
    print(count, ' - ', _2Dpic, ' - ', im.size)
    
    for i in range(18):
        im_rot = im.rotate(21*i, fillcolor='white')
        # im_rot.show()
        im_rot.save("test_data/"+name+"_{}.png".format(i+1))
```

```python
#merging images in rgb mode
import cv2

from os import listdir
from os.path import isfile, join

mypath_r = "test_data/FRONTAL_corrected/"
mypath_g = "test_data/LATERAL/"
mypath_b = "test_data/TOP_correc_rot/"

_2Dfiles_r = [f for f in listdir(mypath_r) if isfile(join(mypath_r, f))]
_2Dfiles_g = [f for f in listdir(mypath_g) if isfile(join(mypath_g, f))]
_2Dfiles_b = [f for f in listdir(mypath_b) if isfile(join(mypath_b, f))]

dsize = (256, 256)

for _2Dpic_r, _2Dpic_g, _2Dpic_b in zip(_2Dfiles_r, _2Dfiles_g, _2Dfiles_b):
    
    input_file_r = mypath_r + _2Dpic_r
    input_file_g = mypath_g + _2Dpic_g
    input_file_b = mypath_b + _2Dpic_b

    name = _2Dpic_r.split('.')[0]
    im_r= cv2.imread(input_file_r)
    im_g= cv2.imread(input_file_g)
    im_b= cv2.imread(input_file_b)
    print(_2Dpic_r, im_r.shape, ' - ', _2Dpic_g, im_g.shape, ' - ', _2Dpic_b, im_b.shape)

    # resize image
    im_r = cv2.resize(im_r, dsize)
    im_g = cv2.resize(im_g, dsize)
    im_b = cv2.resize(im_b, dsize)
    
    #rgb to gray
    im_r = cv2.cvtColor(im_r, cv2.COLOR_BGR2GRAY)
    im_g = cv2.cvtColor(im_g, cv2.COLOR_BGR2GRAY)
    im_b = cv2.cvtColor(im_b, cv2.COLOR_BGR2GRAY)
    
    #merge image into rgb format
    im_rgb = cv2.merge((im_r,im_g,im_b))

    # - save the file as PNG after conversion
    cv2.imwrite("test_data/"+name+"_rgb.png", im_rgb)
```
